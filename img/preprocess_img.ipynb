```python
from PIL import ImageChops, Image

# png2rgb
def png2rgb(img_path):
    
    # white background
    
    rgba_image = Image.open(img_path)
    rgb_im = rgba_image.convert('RGB')
    
    background = Image.new("RGB", rgb_im.size, (255, 255, 255))
    
    # 알파 채널을 이용해 이미지를 배경과 혼합
    rgb_image = Image.alpha_composite(background.convert("RGBA"), rgba_image).convert("RGB")

    return rgb_image

img_path = "gage.png"
rgb_image = png2rgb(img_path)
```

```python
# resize
import os
from PIL import ImageChops, Image

parent_dir = "gif_to_jpg"
size = 360

resize_img_dir = f'{size}_resize_img'

img_dirs = os.listdir(parent_dir)
img_dirs = [ img_dir for img_dir in img_dirs if img_dir not in ['.DS_Store']]

for img_dir in img_dirs:
    
    img_dir_path = os.path.join(parent_dir, img_dir)
    img_nms = os.listdir(img_dir_path)
    img_nms = [ img_nm for img_nm in img_nms if img_nm not in ['.DS_Store']]

    for img_nm in img_nms:
        
        img_src_path = os.path.join(parent_dir, img_dir, img_nm)
        img_dst_path = os.path.join(resize_img_dir, img_dir, img_nm)

        # read img
        img = Image.open(img_src_path)

        # resize img
        img = img.resize((size, size))

        
        # save img
        # 디렉토리 생성
        dst_dir = os.path.join(resize_img_dir, img_dir)
        os.makedirs(dst_dir, exist_ok=True)
        img.save(img_dst_path)
```

```python
# programming으로 누끼따기
import os
from PIL import Image
import numpy as np

dir_nm = "seed1"
imgs = os.listdir(dir_nm)
imgs = [ img_nm for img_nm in imgs if not img_nm in ['.DS_Store', '.ipynb_checkpoints']]
imgs = sorted(imgs)

folder_nm = "seed1"
save_folder_nm = "seed1_alpha"

for img_nm in imgs:
    
    img_path = os.path.join(folder_nm, img_nm)
    
    # 이미지 파일을 열기
    img = Image.open(img_path).convert("RGBA")
    img_array = np.array(img)
    
    # 흰색에 가까운 색상 범위 설정 (조정 가능)
    tolerance = 8  # 흰색과 얼마나 가까운지 판단하는 허용 오차 (0~255 범위 내에서 조정 가능)
    white_rgb = [255, 255, 255]
    
    # 흰색과 가까운 픽셀을 찾기 위한 조건
    # RGB 값이 흰색에 가깝다면, 해당 픽셀의 알파 채널을 0으로 설정
    near_white = np.all(np.abs(img_array[:, :, :3] - white_rgb) <= tolerance, axis=-1)
    
    # 해당하는 픽셀들의 알파 채널을 0으로 설정
    img_array[near_white, 3] = 0
    
    image = Image.fromarray(img_array)

    alpha_img_nm = os.path.join(save_folder_nm, img_nm)
    image.save(alpha_img_nm)
```
